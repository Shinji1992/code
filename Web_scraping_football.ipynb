```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter('ignore')
```

```python
# Defines url where you want to extract data
load_url = "https://www.betexplorer.com/soccer/germany/"
html = requests.get(load_url)
soup = BeautifulSoup(html.content, "html.parser")
```

```python
# Extracts text and new urls at <a href>
results = {}
for i in soup.findAll('a'):
    name = i.get_text()
    url = i.get("href")
    results.setdefault(name, []).append(url)
```

```python
# Creates dataframe from dictionary 
results = pd.DataFrame.from_dict(results, orient='index').T
results.shape
```

```python
# Creates features name list
list_name = list(results)
```

```python
# Extracts only spesific data
list_true_name = []
for i in list_name:
    a = results[i].str.contains('/soccer/germany/',na=False)
    b = any(a)
    if b == True:
        list_true_name.append(i)
```

```python
# Creates dataframe
results = results[list_true_name]
results.shape
```

```python
pd.set_option('display.max_columns', 100)
results.head()
```

```python
# Extracts data which contained football match in 2019. It requires checking which data had the matches in 2019 on the website.
results = results.iloc[:1, 0:36]
results
```

```python
# Below features were different of url format. That is why they were removed. 
list_name.remove('Super Cup')
list_name.remove('Telekom Cup - winter')
```

```python
results_series = results.iloc[0]
results_series
```

```python
# Modifies url
results_series = results_series.str.strip('/')
football_2019_20 = 'https://www.betexplorer.com/'+results_series+'-2019-2020/results/'
football_2018_19 = 'https://www.betexplorer.com/'+results_series+'-2018-2019/results/'
football_2019_20 = football_2019_20[list_name]
football_2018_19 = football_2018_19[list_name]

football_2019 = results_series[['Super Cup','Telekom Cup - winter']]
football_2019 = 'https://www.betexplorer.com/'+football_2019+'-2019/results/'
```

```python
# Extracts data from above urls
def extract_data(url,list_name,dataframe):
    for i,j in zip (url,list_name):
        load_url = i
        html = requests.get(load_url)
        soup = BeautifulSoup(html.content, "html.parser")
    
        for k in soup.find_all(class_='h-text-right h-text-no-wrap'):
            text = k.get_text()
            dataframe.setdefault(j, []).append(text)
    return dataframe
```

```python
df = {}
list_name_2019 = list(['Super Cup','Telekom Cup - winter'])
data_2019 = extract_data(football_2019,list_name_2019,df)
data_2019
```

```python
df = {}
data_2019_20 = extract_data(football_2019_20,list_name,df)
```

```python
df = {}
data_2018_19 = extract_data(football_2018_19,list_name,df)
```

```python
# Creates dataframe
df_2019 = pd.DataFrame.from_dict(data_2019, orient='index')
df_2019_20 = pd.DataFrame.from_dict(data_2019_20, orient='index')
df_2018_19 = pd.DataFrame.from_dict(data_2018_19, orient='index')
```

```python
# Cleaning dataframe by counting values and merging dataframes
a = df_2019.T
b1 = a['Super Cup'].value_counts()
b2 = a['Telekom Cup - winter'].value_counts()

b1_1 = pd.DataFrame(data=b1)
b2_1 = pd.DataFrame(data=b2)

b1_1['date'] = b1_1.index
b1_1 = b1_1.reset_index().drop("index", axis=1)
b2_1['date'] = b2_1.index
b2_1 = b2_1.reset_index().drop("index", axis=1)

final = pd.merge(b1_1,b2_1, how='outer')
final = final[['date','Super Cup','Telekom Cup - winter']]
final
```

```python
# Counts values and merges all daraframes
def count(data,name,dataframe):
    a = data.T
    check = data.T.columns.values.tolist()
    for i in check:
        if i in name:
            b = a[i].value_counts()
            final = pd.DataFrame(data=b)
            final['date'] = final.index
            final = final.reset_index().drop("index", axis=1)
            dataframe = pd.merge(dataframe,final, how='outer')
        pass
    return dataframe
```

```python
df_2019_20_count = count(df_2019_20,list_name,final)
df_2019_20_count
```

```python
all_data = count(df_2018_19,list_name,df_2019_20_count)
```

```python
# Extracts data in 2019
all_data_2019=all_data[all_data['date'].str.contains('2019')]
all_data_2019.shape
```

```python
# Sums up all matches at same day
a = all_data_2019.sum(axis=1)
all_data_2019['sum'] = a
```

```python
all_data_2019
```

```python
new_df = all_data_2019[['date','sum']]
new_df
```

```python
# Groups by date, and sums up the matches
final = new_df.groupby("date")["sum"].sum()
final
```

```python
# Converts to dataframe
new = pd.DataFrame(data=final)
new['Date'] = new.index
new = final.reset_index()
new
```

```python
new_df = new.rename(columns = {"sum": "Football_count"})
```

```python
# Modifies date
new_df = pd.concat([new_df, new_df['date'].str.split('.', expand=True)], axis=1).drop('date', axis=1)
new_df.head()
```

```python
# Modifies date
final = pd.concat([new_df[2]+'-'+new_df[1]+'-'+new_df[0],new_df['Football_count']],axis=1)
final['date']=final[0]
final = final.drop(0, axis=1)
final
```

```python
# Extract professional football games
def extract_data(url_list,name,dataframe):
    for i in (url_list):
        html = requests.get(i)
        soup = BeautifulSoup(html.content, "html.parser")
    
        for k in soup.find_all(class_='h-text-right h-text-no-wrap'):
            text = k.get_text()
            dataframe.setdefault(name, []).append(text)
    return dataframe
```

```python
bundesliga_list = list(['https://www.betexplorer.com/soccer/germany/bundesliga-2019-2020/results/',
                        'https://www.betexplorer.com/soccer/germany/bundesliga-2018-2019/results/'])

df = {}
Bundesliga_data = extract_data(bundesliga_list,'Bundesliga',df)
```

```python
bundesliga2_list = list(['https://www.betexplorer.com/soccer/germany/2-bundesliga-2019-2020/results/',
                        'https://www.betexplorer.com/soccer/germany/2-bundesliga-2018-2019/results/'])

df = {}
Bundesliga2_data = extract_data(bundesliga2_list,'2.Bundesliga',df)
```

```python
liga3_list = list(['https://www.betexplorer.com/soccer/germany/3-liga-2019-2020/results/',
                   'https://www.betexplorer.com/soccer/germany/3-liga-2018-2019/results/'])

df = {}
Liga3_data = extract_data(liga3_list,'3.liga',df)
```

```python
Bundesliga_df = pd.DataFrame.from_dict(Bundesliga_data, orient='index')
Bundesliga2_df = pd.DataFrame.from_dict(Bundesliga2_data, orient='index')
Liga3_df = pd.DataFrame.from_dict(Liga3_data, orient='index')
```

```python
def count(data,name):
    a = data.T
    b = a[name].value_counts()
    final = pd.DataFrame(data=b)
    final['date'] = final.index
    final = final.reset_index().drop("index", axis=1)
    return final
```

```python
Bundesliga_df = count(Bundesliga_df,'Bundesliga')
Bundesliga2_df = count(Bundesliga2_df,'2.Bundesliga')
Liga3_df = count(Liga3_df,'3.liga')
```

```python
all_df = pd.merge(Bundesliga_df,Bundesliga2_df, how='outer')
all_df = pd.merge(all_df,Liga3_df, how='outer')
all_df
```

```python
all_df = all_df[(all_df['date'].str.contains('2019'))]
all_df
```

```python
new_df = pd.concat([all_df, all_df['date'].str.split('.', expand=True)], axis=1).drop('date', axis=1)
new_df
```

```python
# Modifies date
date = pd.concat([new_df[2]+'-'+new_df[1]+'-'+new_df[0]],axis=1)
date['date']=date[0]
final = pd.concat([new_df,date],axis=1)
final = final.drop(0, axis=1)
final = final.drop(1, axis=1)
final = final.drop(2, axis=1)
final
```

```python
# Saves data as csv file
final.to_csv("No_pro_football_match.csv")
```
